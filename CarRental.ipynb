# **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import *
```

# **Import Dataset**

```python
dataset=pd.read_csv('CarRentalData.csv')
dataset.head()
```

**To Check for the null values**

```python
dataset.isnull().sum()
```

# **No Of Rows And Columns**

# **Description of Data**

```python
dataset.describe()
```

**Dealing with the Null data**

```python
dataset['fuelType'].fillna(method ='ffill',  inplace = True)
dataset['rating'].fillna(method ='ffill',  inplace = True)
dataset.isnull().sum()
```

# Encoding IndeDependent Variable

# **Encoding The Dependent Variable**

```python
dataset['type'].value_counts()
```

```python
dataset.replace({'type':{'sedan':0,'suv':1,'minivan':2,'truck':3,'van':4},'country':{'US':1}},inplace=True)
dataset.head()
```

```python
dataset['type'].value_counts()
```

```python
x=dataset.drop(['type','latitude','longitude','id'],axis=1)
y=dataset['type']
```

```python
pd.DataFrame(x)
```

```python
pd.DataFrame(y)
```

```python
x = pd.DataFrame(x)
for column in x.columns:
  print(column)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(dtype=np.int32,sparse=False),[0,4,6,8,9])],remainder='passthrough')
x=ct.fit_transform(x)
```
