# **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import *
```

# **Import Dataset**

```python
dataset=pd.read_csv('CarRentalData.csv')
dataset.head()
```

**To Check for the null values**

```python
dataset.isnull().sum()
```

# **No Of Rows And Columns**

```python
dataset.shape
```

# **Description of Data**

```python
dataset.describe()
```

**Dealing with the Null data**

```python
dataset['fuelType'].fillna(method ='ffill',  inplace = True)
dataset['rating'].fillna(method ='ffill',  inplace = True)
dataset.isnull().sum()
```

# Encoding IndeDependent Variable

# **Encoding The Dependent Variable**

```python
dataset['type'].value_counts()
```

```python
dataset.replace({'type':{'sedan':0,'suv':1,'minivan':2,'truck':3,'van':4},'country':{'US':1}},inplace=True)
dataset.head()
```

```python
dataset['type'].value_counts()
```

```python
x=dataset.drop(['type','latitude','longitude','id'],axis=1)
y=dataset['type']
```

```python
pd.DataFrame(x)
```

```python
x = pd.DataFrame(x)
for column in x.columns:
  print(column)
```

```python
pd.DataFrame(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(dtype=np.int32,sparse=False),[0,4,6,8,9])],remainder='passthrough')
x=ct.fit_transform(x)
```

```python
pd.DataFrame(x)
```

# **Split Test and Train data**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
pd.DataFrame(X_train)
```

```python
pd.DataFrame(X_test)
```

# **Feature scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
np.asarray(X_test[0])
```

# **Training The svc Model**

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

# **Predicting of The Test Set**

```python
y_pred=classifier.predict(X_test)
```

# **Confusion Metrics and accuracy score**

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
acc=accuracy_score(y_test, y_pred)
print('The Accuracy for the Model will be :',acc*100,'%')
```

# **Making A Predictive System**

```python
n=int(input("Enter the Number of Data Upto which u want to check the data in X_test"))
for a in range(0,n+1):
  inp=X_test[a].reshape(1,-1)
  prediction=classifier.predict(inp)
  print(prediction)
  if(prediction==0):
    print("The Vehicle Rented is a Car as it is mapped with value ",prediction)
  elif(prediction==1):
    print("The Vehicle Rented is a SUV as it is mapped with value ",prediction)
  elif(prediction==2):
    print("The Vehicle Rented is a minivan as it is mapped with value ",prediction)
  elif(prediction==3):
    print("The Vehicle Rented is a truck as it is mapped with value ",prediction)
  elif(prediction==4):
    print("The Vehicle Rented is a van as it is mapped with value ",prediction)
  else:
    print('Unable to Predict')
```
